```python
# 미세먼지 안녕 - 백준

R, C, T = list(map(int, input().split()))
mat = [[0 for i in range(C)] for j in range(R)]

gong_pose = []

for i in range(R):
    mat[i] = list(map(int, input().split()))
    for j in range(C):
        if(mat[i][j] == -1):
            gong_pose.append([i,j])

# print(gong_pose)

dys = [0,-1,0,1]
dxs = [1,0,-1,0]

def in_range(y,x):
    return 0 <= y and y< R and 0<=x and x<C


for time in range(T):


    ### 미세먼지 확산
    # 각 rc 자리의 네 방향 확인해서 값 옮기기 (in range, 공청 pose 확인)
    tem_mat = [[0 for i in range(C)] for j in range(R)]

    # 업데이트할 값들 구하기
    for r in range(R):
        for c in range(C):
            diffu_val = int(mat[r][c]/5)
            for dy, dx in zip(dys, dxs):
                ny = r+dy
                nx = c+dx
                if(not in_range(ny,nx) or [ny,nx] in gong_pose):
                    continue

                tem_mat[ny][nx] += diffu_val
                tem_mat[r][c] -= diffu_val

    # 동시에 업데이트
    for r in range(R):
        for c in range(C):
            mat[r][c] += tem_mat[r][c]

    # print()
    # for r in range(R):
    #     for c in range(C):
    #         print(mat[r][c], end=' ')
    #     print()





    ### 공기청정기 순환

    for idx in range(len(gong_pose)):
        y,x = gong_pose[idx]

        if (idx == 0): # 위로
            cur_y, cur_x = y - 1, x
            dy, dx = -1, 0

            while True:
                ny, nx = cur_y + dy, cur_x + dx
                mat[cur_y][cur_x] = mat[ny][nx]
                cur_y, cur_x = ny, nx

                if(ny == 0 and nx == 0):
                    dy, dx = 0, 1
                elif(ny == 0 and nx == C-1):
                    dy, dx = 1, 0
                elif(ny ==  y and nx == C-1):
                    dy, dx = 0, -1
                elif(ny == y and nx == x+1):
                    mat[ny][nx] = 0
                    break

        elif (idx == 1): # 아래로
            cur_y, cur_x = y + 1, x
            dy, dx = 1, 0

            while True:
                ny, nx = cur_y + dy, cur_x + dx
                mat[cur_y][cur_x] = mat[ny][nx]
                cur_y, cur_x = ny, nx

                if(ny == R-1 and nx == 0):
                    dy, dx = 0, 1
                elif(ny == R-1 and nx == C-1):
                    dy, dx = -1, 0
                elif(ny ==  y and nx == C-1):
                    dy, dx = 0, -1
                elif(ny == y and nx == x+1):
                    mat[ny][nx] = 0
                    break


    # print()
    # for r in range(R):
    #     for c in range(C):
    #         print(mat[r][c], end=' ')
    #     print()



total = 0
for i in range(R):
    total += sum(mat[i])

print(total + 2)
```
