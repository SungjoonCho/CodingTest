```python
# 쿼드압축 후 개수 세기 - 프로그래머스
# https://school.programmers.co.kr/learn/courses/30/lessons/68936

# 4분할하고 queue에 삽입
# 4개 조각 확인해서 통합되면 1,0 리스트에 +1
# 통합 안되면 4분할 해서 queue에 삽입

def sameCheck(arr, r_s, r_e, c_s, c_e):

    init = arr[int(r_s)][int(c_s)]

    for i in range(int(r_s), int(r_e)):
        for j in range(int(c_s), int(c_e)):
            if(init != arr[i][j]):
                return False, -1

    return True, init

def solution(arr):
    answer = []

    N = len(arr)
    divided_N = int(N/2)
    queue = [[0,N,0,N]]

    numList = [0,0] # 0 cnt , 1 cnt

    while(queue):

        # r시작 r끝 c시작 c끝
        r_s, r_e, c_s, c_e = queue.pop(0)

        # 통합 확인
        res, num = sameCheck(arr, r_s, r_e, c_s, c_e)

        # 통합되면
        if(res == True):
            numList[num] += 1
            continue


        # 통합 안되면 4분할 해서 queue에 삽입
        diff = r_e - r_s

        # 그런데 더 이상 쪼개지지 않으면 삽입 X => 할 필요 없음 (위에서 res True 나와서 어차피 여기 안 옴)
        # if (diff==1):
        #     continue

        diff /= 2

        queue.append([r_s,r_s+diff,c_s,c_s+diff])
        queue.append([r_s,r_s+diff,c_s + diff, c_e])
        queue.append([r_s+diff,r_e,c_s,c_s+diff])
        queue.append([r_s+diff,r_e,c_s + diff, c_e])


    return numList
```
