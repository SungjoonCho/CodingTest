```python

# 큰 수 만들기 - 프로그래머스
# https://school.programmers.co.kr/learn/courses/30/lessons/42883

def solution(number, k):

    # 0번째 넣어놓고 시작
    # queue에 최종적으로 남은 숫자들이 답이 될거임
    queue = [number[0]]
    cnt = 0
    for i in range(1, len(number)):
        num = number[i]

        # queue에 있는 마지막 숫자가 현재 숫자보다 작으면 pop (작지 않은 숫자가 나올 때까지 모두 pop)
        while(queue and queue[-1] < num and cnt < k):
            queue.pop()
            cnt += 1

        # 현재 숫자 넣어주기 (만약에 cnt가 위의 while문에서 k가 되면 조건 만족이므로 queue에 남은 것들 모두 넣어줌)
        queue.append(num)

    # k 만족 안됐는데 for문 끝나면 queue에 있는 것들 중에 마지막 숫자들 pop
    while(cnt < k):
        queue.pop()
        cnt += 1

    return "".join(queue)
```
