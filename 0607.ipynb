```python
# 격자숫자놀이 - 코드트리
# https://www.codetree.ai/training-field/frequent-problems/problems/matrix-number-play/description?page=3&pageSize=20


r, c, k = list(map(int, input().split()))
r = r - 1
c = c - 1

mat = [[0 for i in range(3)] for j in range(3)]

for i in range(3):
    mat[i] = list(map(int, input().split()))


def in_range(y, x):
    return 0 <= y and y < len(mat) and 0 <= x and x < len(mat[0])


time = 0
# print(r,c)

# for i in mat:
#     print(i)
# print()

while (True):

    if (in_range(r, c) and mat[r][c] == k):
        print(time)
        break

    time += 1

    # print(time)

    if (time > 100):
        print(-1)
        break

    row_length = len(mat)
    col_length = len(mat[0])

    # 행의 개수가 열의 개수보다 크거나 같은 경우
    if (row_length >= col_length):

        maxLenth = 0
        for r_idx in range(row_length):

            cntDict = {}
            for c_idx in range(col_length):
                num = mat[r_idx][c_idx]
                if (num == 0):
                    continue

                if (num not in cntDict.keys()):
                    cntDict[num] = 1
                else:
                    cntDict[num] += 1

            # 정렬
            numPairs = []
            for key, value in cntDict.items():
                numPairs.append([key, value])
            numPairs.sort(key=lambda x: (x[1], x[0]))

            # 하나의 리스트롤 만들기
            totalList = []
            for each_pair in numPairs:
                totalList += each_pair

            mat[r_idx] = totalList[:min(len(totalList), 100)]
            if (len(totalList) > maxLenth):
                maxLenth = min(len(totalList), 100)

        # 빈 공간 0으로 채워주기
        for rIdx in range(len(mat)):
            while (len(mat[rIdx]) < maxLenth):
                mat[rIdx].append(0)

        # print('case1')
        # for i in mat:
        #     print(i)
        # print()


    else:  # 행의 개수가 열의 개수보다 작은 경우

        temList = []
        maxLenth = 0
        for c_idx in range(col_length):

            cntDict = {}
            for r_idx in range(row_length):
                num = mat[r_idx][c_idx]
                if (num == 0):
                    continue

                if (num not in cntDict.keys()):
                    cntDict[num] = 1
                else:
                    cntDict[num] += 1
            # print(cntDict)

            # 정렬
            numPairs = []
            for key, value in cntDict.items():
                numPairs.append([key, value])
            numPairs.sort(key=lambda x: (x[1], x[0]))

            # 하나의 리스트로 만들기
            totalList = []
            for each_pair in numPairs:
                totalList += each_pair

            # 대입
            temList.append(totalList[:min(len(totalList), 100)])
            # print(totalList)

            # 최대 길이 정하기
            if (len(totalList) > maxLenth):
                maxLenth = min(len(totalList), 100) # 이게 행 길이가 될거임

        # 빈공간 0으로 채워주기
        for rIdx in range(len(temList)):
            while (len(temList[rIdx]) < maxLenth):
                temList[rIdx].append(0)

        # transpose
        mat = [[0 for i in range(len(temList))] for j in range(len(temList[0]))]
        for rIdx in range(len(temList[0])):
            for cIdx in range(len(temList)):
                mat[rIdx][cIdx] = temList[cIdx][rIdx]


        # print('case2')
        # for i in mat:
        #     print(i)
        # print()




```
