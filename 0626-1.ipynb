```python
# N으로 표현 (프로그래머스)

def solution(N, number):
    answer = -1


    # 각 리스트의 경우의 수 초기값 (숫자만 이어붙인 경우)
    caseList = [[] for i in range(8)]
    for i in range(8):
        val = int(str(N) * (i+1))
        if(val == number):
            return i+1
        caseList[i].append(int(str(N) * (i+1)))




    for i in range(1,8):

        # case i (i+1개 숫자 활용) 자리에 들어가기 위해서는 j, k의 조합 필요 (ex. 5개 숫자를 활용하려면 1,4 / 2,3 자리의 dp 데이터 가져오기)
        for j in range(0, i//2+1):
            for k in range(0, i):
                if(j+k != i-1 or j>k):
                    continue
                # print(i, j, k)

                source = j
                target = k


                # 이전에 저장된 데이터 가져오기
                sourceList = caseList[source][:]
                targetList = caseList[target][:]

                # 사칙연산 (N x M 개 모든 조합에 대해 사칙연산 해보기)
                for t_eachVal in targetList:
                    for s_eachVal in sourceList:
                        caseList[i].append(t_eachVal + s_eachVal)
                        if(t_eachVal + s_eachVal == number):
                            # print(caseList)
                            return i+1

                        caseList[i].append(t_eachVal - s_eachVal)
                        if(t_eachVal - s_eachVal == number):
                            # print(caseList)
                            return i+1

                        caseList[i].append(s_eachVal - t_eachVal)
                        if(s_eachVal - t_eachVal == number):
                            # print(caseList)
                            return i+1

                        caseList[i].append(t_eachVal * s_eachVal)
                        if(t_eachVal * s_eachVal == number):
                            # print(caseList)
                            return i+1

                        if(s_eachVal != 0):
                            caseList[i].append(t_eachVal // s_eachVal)
                            if(t_eachVal // s_eachVal == number):
                                # print(caseList[i])
                                return i+1

                        if(t_eachVal != 0):
                            caseList[i].append(s_eachVal // t_eachVal)
                            if(s_eachVal // t_eachVal == number):
                                # print(caseList[i])
                                return i+1

        # 양 너무 많을 수 있어서 중복 제거
        caseList[i] = list(set(caseList[i]))


    return answer
```
