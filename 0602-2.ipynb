```python
# 삼각 달팽이 - 프로그래머스
# https://school.programmers.co.kr/learn/courses/30/lessons/68645

def solution(n):

    # 피라미드 mat 초기화
    mat = []
    totalCnt = 0
    for i in range(1,n+1):
        mat.append([0] * i)
        totalCnt = totalCnt + i


    cy, cx = 0, 0
    curDir = 0
    # 하 우 상 좌 (상으로 갈 때만 주의)
    dys = [1,0,-1,0]
    dxs = [0,1,-1,-1]

    # 현재 cy, cx의 값 mat에 업데이트
    mat[cy][cx] = 1
    totalCnt -= 1
    curNum = 2

    def in_range(y,x):
        return 0<=y and y<n and 0<=x and x<n

    while True:

        if(totalCnt == 0):
            break

        # 새로운 cy, cx 계산해서 만약 in range 벗어나거나 값이 있으면 방향 전환
        # cy, cx 업데이트
        while(True):
            ny, nx = cy+dys[curDir], cx+dxs[curDir]
            if(not in_range(ny, nx) or mat[ny][nx] != 0):
                curDir = (curDir+1) % 4
            else:
                break


        # cy, cx에 해당하는 값 mat에 업데이트
        cy, cx = cy+dys[curDir], cx+dxs[curDir]
        mat[cy][cx] = curNum

        curNum += 1
        totalCnt -= 1

    # answer에 구한 mat의 각 row 합치기
    answer = []
    for each in mat:
        answer += each

    return answer
```
