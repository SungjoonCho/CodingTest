```python
# 설탕 배달 - 백준

n = int(input())

dp1 = [0,0,0,0,0,1]
dp2 = [0,0,0,1,0,0]
dp_final = [0,0,0,1,0,1]

if(n <= 5):
    result = dp_final[n]
    if(result == 0):
        print(-1)
    else:
        print(result)
    exit()

## 현재값 i에서 5,3 각각을 빼봐서 그 때의 횟수에 +1 해주는 방식임
# dp_final은 각 i를 만들기 위한 최소 횟수 테이블
# dp1 은 i에서 5 뺀 경우 + 1
# dp2 은 i에서 3 뺀 경우 + 1
# 그리고 dp1 dp2 둘 중 최소값이 현재 dp_final의 i에 들어감

for i in range(6, n+1):
    val1 = dp_final[i-5]
    val2 = dp_final[i-3]

    if(val1 == 0):
        val1 = 1000000
    else:
        val1 += 1

    if(val2 == 0):
        val2 = 1000000
    else:
        val2 += 1

    final_val = min(val1, val2)
    if(final_val == 1000000):
        dp_final.append(0)
    else:
        dp_final.append(min(val1, val2))

# print(dp_final)

if(dp_final[n] == 0):
    print(-1)
else:
    print(dp_final[n])
```
