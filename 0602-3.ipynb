```python
# 풍선 터뜨리기 - 프로그래머스
# 시간 초과..리스트 길이가 1,000,000 이하여서 새로 짜야할듯

finalList = []
temList = []

def selectPair(tfList, pair):
    start = pair[0]
    for i in range(start+1, len(tfList)):
        if(tfList[i] == True):
            return True, pair + [i]

    return False, pair


# smallBreak은 번호가 더 작은 풍선 터트렸는지 여부
def dfs(tfList, T_len, smallBreak):
    # print()
    # print(tfList, T_len, smallBreak)
    global finalList

    # TF의 T len이 1이면 return (finalList에 append)
    if(T_len == 1):
        # print(T_len)
        for i in range(len(tfList)):
            if(tfList[i] == True):
                finalList.append(temList[i])
                # print(i)
                return


    #0,1 1,2 2,3 ~~ 이렇게 선택 (while문으로 연속되는 쌍이 있을때까지)
    # T T F T

    # 맨 처음 True인 애 찾기
    idx = 0
    for i in range(len(tfList)):
        if(tfList[i] == True):
            idx = i
            break

    pair = [idx]

    while True: # 리턴 값이 False면 끝 (pair 못 찾으면 끝임)
        res, pair = selectPair(tfList[:], pair[:])
        # print(pair, res)
        if res == False:
            # print(pair, res)
            break

        # 해당 pair에 대해서 둘 중 하나 각 케이스 tfList False 만들고, 더작은 풍선 터트릴지 고려
            # 각각에 대해 dfs 진행
        if(temList[pair[0]] > temList[pair[1]]):
            if(smallBreak == False):
                tfList[pair[1]] = False # 더 작은 것에 대해서 터트리기
                dfs(tfList[:], T_len-1, True)
                tfList[pair[1]] = True

            tfList[pair[0]] = False # 큰 것 터뜨리기
            dfs(tfList[:], T_len-1, smallBreak)
            tfList[pair[0]] = True

        else:
            if(smallBreak == False):
                tfList[pair[0]] = False # 더 작은 것에 대해서 터트리기
                dfs(tfList[:], T_len-1, True)
                tfList[pair[0]] = True

            tfList[pair[1]] = False
            dfs(tfList[:], T_len-1, smallBreak)
            tfList[pair[1]] = True


        pair = [pair[1]]





def solution(a):
    global temList
    temList = a[:]

    answer = 0

    tfList = [True for i in range(len(temList))]

    dfs(tfList[:], len(a), False)

    return len(set(finalList))
```
