```python

# https://www.acmicpc.net/problem/13334
# 백준 - 철로

# 리스트로 안 되고 heapq로 해야됨

import heapq


personNum = int(input())
personList = []

for i in range(personNum):
    personList.append(sorted(list(map(int, input().split()))))

d = int(input())

# 각 사람의 끝지점을 기준으로 오름차순 정렬
personList.sort(key=lambda x : (x[1], x[0]))
# print(personList)

curMax = 0

queue = [] # 각 사람의 시작점 저장 (나중에 queue의 시작점을 벗어나는 애들은 pop하려고)
for idx in range(len(personList)):
    start, end = personList[idx]

    if(end - start <= d):
        # queue.append(start)
        heapq.heappush(queue, start)
    else:
        continue

    while(queue):
        if(end - queue[0] <= d):
            break
        # queue.pop(0)
        heapq.heappop(queue)

    # print(queue)
    curMax = max(len(queue), curMax)

print(curMax)
```
