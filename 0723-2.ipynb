```python
# 프로그래머스(도둑질)


def solution(money):
	answer = []
	# 첫번째 포함하는 경우
	dp1 = []
	dp2 = []

	dp1.append(money[0])
	dp2.append(0)
	dp1.append(money[1])
	dp2.append(dp1[0])

	for i in range(2, len(money)):
	    if(i == len(money)-1):
	        dp2.append(max(dp1[i-1], dp1[i-2]))
	    else:
	        dp1.append(max([dp2[i-1], dp1[i-2], dp2[i-2]]) + money[i])
	        dp2.append(max(dp1[i-1], dp1[i-2]))
	answer.append(max(dp1[-1], dp2[-1]))

	# 첫번째 포함 안 하는 경우
	dp1 = []
	dp2 = []

	dp1.append(0)
	dp2.append(0)
	dp1.append(money[1])
	dp2.append(0)

	for i in range(2, len(money)):
	    if(i == len(money)-1):
	        dp1.append(max([dp2[i-1], dp1[i-2], dp2[i-2]]) + money[i])
	        dp2.append(max(dp1[i-1], dp1[i-2]))
	    else:
	        dp1.append(max([dp2[i-1], dp1[i-2], dp2[i-2]]) + money[i])
	        dp2.append(max(dp1[i-1], dp1[i-2]))
	answer.append(max(dp1[-1], dp2[-1]))

	return max(answer)
```
