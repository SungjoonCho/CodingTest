```python
# 문자열 폭발
# https://www.acmicpc.net/problem/9935

queue = []
inputStr = input()
removeStr = input()

curIdx = 0
while (len(inputStr) > curIdx):
    #
    # print(queue)
    # print()

    curAlpha = inputStr[curIdx]
    queue.append(curAlpha)

    curIdx += 1

    if (len(queue) < len(removeStr)):
        continue

    if("".join(queue[len(queue) - len(removeStr):]) == removeStr):
        for _ in range(len(removeStr)):
            queue.pop()

if (len(queue) == 0):
    print("FRULA")
else:
    print("".join(queue))
```
