```python
# 프로그래머스 - 이모티콘 할인행사

def solution(users, emoticons):
    caseList = [] # 모든 경우의수 담기 ([임티플러스 가입자 수, 판매액])
    priceRatio = [10, 20, 30, 40]

    curList = [0 for i in range(len(emoticons))]
    while curList[0] != len(priceRatio):

        # 변경된 가격, 할인율 저장
        changedPrice = []
        curRatio = []
        for idx in range(len(emoticons)):
            discValue = priceRatio[curList[idx]]
            curRatio.append(discValue)
            changedPrice.append(emoticons[idx] * (100-discValue) * 0.01)
        # print(curRatio, changedPrice)

        # 플러스 가입자 수, 가격 계산 후 caseList에 추가
        totalVal = 0
        total_plus = 0
        for p_idx in range(len(users)):
            discThres = users[p_idx][0]
            moneyThres = users[p_idx][1]

            valSum = 0
            emo_plus = 0
            # print(discThres, moneyThres)
            for e_idx in range(len(emoticons)):
                if(discThres <= curRatio[e_idx]):
                    valSum += changedPrice[e_idx]

            if(valSum >= moneyThres):
                emo_plus = 1
                valSum = 0

            # print(emo_plus, valSum)

            totalVal += valSum
            total_plus += emo_plus

        caseList.append([total_plus, totalVal])



        # 1 더하기
        # 0000 0001 0002 0003 0010 0011 0012 0013 0020 ... 0033 0100
        # caseList.append(curList[:])
        curList[-1] += 1

        for i in range(len(curList)-1, -1, -1):
            if(curList[i] != len(priceRatio)):
                break

            # 끝난 경우
            if(i == 0 and curList[0] == len(priceRatio)):
                break

            # 다음 자리 올림
            curList[i] = 0
            curList[i-1] += 1

    # print(caseList)


    caseList.sort(reverse=True)

    return caseList[0]


```
