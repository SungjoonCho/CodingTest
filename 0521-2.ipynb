```python
# 미로 탈출 - 프로그래머스


def in_range(y,x, max_y, max_x):
    return 0<=y and y<max_y and 0<=x and x<max_x

def solution(maps):
    answer = 0

    start = []
    lever = []
    end = []
    for r in range(len(maps)):
        for c in range(len(maps[0])):
            if(maps[r][c] == 'S'):
                start = [r,c]
            elif(maps[r][c] == 'L'):
                lever = [r,c]
            elif(maps[r][c] == 'E'):
                end = [r,c]


    ## start ~ lever까지 bfs
    visited = [[False for i in range(len(maps[0]))] for j in range(len(maps))]
    bfs_mat = [[0 for i in range(len(maps[0]))] for j in range(len(maps))]
    queue = [start]
    visited[start[0]][start[1]] = True

    dys = [0,-1,0,1]
    dxs = [1,0,-1,0]

    while(queue and visited[lever[0]][lever[1]] == False):

        cy, cx = queue.pop(0)
        for dy, dx in zip(dys, dxs):
            ny = cy + dy
            nx = cx + dx
            if(not in_range(ny, nx, len(maps), len(maps[0])) or maps[ny][nx] == 'X' or visited[ny][nx] == True):
                continue

            queue.append([ny,nx])
            visited[ny][nx] = True
            bfs_mat[ny][nx] = bfs_mat[cy][cx] + 1

    if(visited[lever[0]][lever[1]] == False):
        return -1

    answer += bfs_mat[lever[0]][lever[1]]
    # print(answer)

    ## lever ~ exit까지 bfs
    visited = [[False for i in range(len(maps[0]))] for j in range(len(maps))]
    bfs_mat = [[0 for i in range(len(maps[0]))] for j in range(len(maps))]
    queue = [lever]
    visited[lever[0]][lever[1]] = True

    dys = [0,-1,0,1]
    dxs = [1,0,-1,0]

    while(queue and visited[end[0]][end[1]] == False):

        cy, cx = queue.pop(0)
        for dy, dx in zip(dys, dxs):
            ny = cy + dy
            nx = cx + dx
            if(not in_range(ny, nx, len(maps), len(maps[0])) or maps[ny][nx] == 'X' or visited[ny][nx] == True):
                continue

            queue.append([ny,nx])
            visited[ny][nx] = True
            bfs_mat[ny][nx] = bfs_mat[cy][cx] + 1



    if(visited[end[0]][end[1]] == False):
        return -1

    answer += bfs_mat[end[0]][end[1]]

    return answer
```
