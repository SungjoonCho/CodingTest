```python
# 프로그래머스 - 주차 요금 계산

def convertTime(time):
    hour = int(time[:2])
    minute = int(time[3:])

    return hour*60 + minute



def solution(fees, records):
    answer = []
    giTime, giFee, danTime, danFee = fees


    # 시간 누적
    carDict = {} # key : carNum value : index
    carList = [] # [carNum, 요금, inout]
    for each_info in records:
        word = each_info.split()
        time, carNum, inout = word
        time = convertTime(time)

        # 새로 in
        if(carNum not in carDict.keys()):
            # print(carNum, 'in')
            carDict[carNum] = len(carList)
            carList.append([carNum, 0, time, 1, 0])
            continue

        carIdx = carDict[carNum]
        # in 기록이 있는 경우 out 처리
        if carList[carIdx][3] == 1:
            # print(carList[carIdx][0], 'out')
            timeDiff = time - carList[carIdx][2]
            carList[carIdx][4] += timeDiff # 시간 누적
            carList[carIdx][3] = 0 # out

        # out인데 다시 온 경우
        elif carList[carIdx][3] == 0:
            # print(carList[carIdx][0], 'in')
            carList[carIdx][2] = time
            carList[carIdx][3] = 1


    # 비용 계산
    for carIdx in range(len(carList)):
        if(carList[carIdx][3] == 1): # 아직 안나간 경우
            # print(carList[carIdx][0])
            timeDiff = convertTime("23:59") - carList[carIdx][2]
            carList[carIdx][4] += timeDiff

        timeDiff = carList[carIdx][4]
        if(timeDiff > giTime):
            timeCnt = 0
            timeDiff -= giTime
            carList[carIdx][1] += giFee
            if(timeDiff % danTime != 0):
                timeCnt = int(timeDiff/danTime) + 1
            else:
                timeCnt = int(timeDiff/danTime)
            carList[carIdx][1] += timeCnt * danFee
        else:
            carList[carIdx][1] += giFee


    # print(carList)
    carList.sort(key= lambda x:(x[0]))

    answer = []
    for i in carList:
        answer.append(i[1])

    return answer
```
