```python
# 함께하는 효도 (Softeer)
# https://softeer.ai/practice/7727

n, m = list(map(int, input().split()))

mat = [[0 for i in range(n)] for j in range(n)]
for r in range(n):
    mat[r] = list(map(int, input().split()))

peopleList = []
for p in range(m):
    tem = list(map(int, input().split()))
    peopleList.append([each-1 for each in tem])


dys = [0,-1,0,1]
dxs = [1,0,-1,0]

totalRes = 0
visitedMat = [[-1 for i in range(n)] for j in range(n)]
for p in range(m):
    y, x = peopleList[p]
    visitedMat[y][x] = p
    totalRes += mat[y][x]

maxNum = 0

def in_range(y,x):
    return 0<=y and y<n and 0<=x and x<n

def dfs(pIdx, curCoord, cnt, result):
    global maxNum, visitedMat
    # 종료 조건

    if(cnt == 3):
        cnt = 1
        pIdx += 1

        if(pIdx == len(peopleList)):
            # 합계에 추가
            if(result > maxNum):
                maxNum = result
                # print(maxNum)
                # for r in visitedMat:
                #     print(r)
                # print()
            return
        else:
            curCoord = peopleList[pIdx][:]

    else:
        cnt += 1


    cy, cx = curCoord
    # 네 방향 go
    for dy, dx in zip(dys, dxs):
        ny, nx = cy+dy, cx+dx

        if(not in_range(ny, nx) or visitedMat[ny][nx] > -1):
            continue

        visitedMat[ny][nx] = pIdx

        dfs(pIdx,
            [ny,nx],
            cnt,
            result + mat[ny][nx])

        visitedMat[ny][nx] = -1




# pIdx, curCoord, cnt, result


dfs(0, peopleList[0][:], 0, totalRes)

print(maxNum)
```
