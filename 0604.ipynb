```python
# 불안한 무빙워크 - 코드트리
# https://www.codetree.ai/training-field/frequent-problems/problems/unstable-moving-walk/description?page=2&pageSize=20


n, k = list(map(int, input().split()))
stableList = list(map(int, input().split()))
personList = []

# stableList를 queue처럼 써야 할듯
    # 각 칸의 안정성 저장
# personList에 신규는 insert(0, 0)
    # 각 사람의 pose 저장


# 무빙워크 한칸 이동
    # stableList pop() insert(0)
    # personList 길이가 0보다 크면 다들 한칸씩 이동
        # 만약 그 사람이 n보다 크면 아웃

# personList 길이가 0보다 클 때 각 사람에 대해서 앞선 칸에 사람이 없고 앞선 칸의 안정성이 0이 아니면:
    # 각 사람들 한칸씩 이동 - # 안정성 1 감소
        # 만약 그 사람이 n보다 크면 아웃

# 0번 무빙워크 칸에 사람 없고 안정성 0 아니면 personList에 사람 추가 (0 추가)
    # 안정성 1 감소

# 안정성이 0인 칸이 k개 이상이라면 과정을 종료

turn = 0
keepGo = True
while keepGo:
    turn += 1

    # 무빙워크 한칸 이동
    last = stableList.pop()
    stableList.insert(0, last)

    print('aft movingwalk step')
    print('st', stableList)
    print('pe', personList)
    print()

    # personList 길이가 0보다 크면 다들 한칸씩 이동
        # 만약 그 사람이 n보다 크면 아웃
    personLen = len(personList)
    for eachIdx in range(personLen-1, -1, -1):
        personList[eachIdx] += 1
        if personList[eachIdx] > n-1:
            personList.pop()

    print('aft person auto move')
    print('st', stableList)
    print('pe', personList)
    print()


    # personList 길이가 0보다 클 때 각 사람에 대해서 앞선 칸에 사람이 없고 앞선 칸의 안정성이 0이 아니면:
    # 각 사람들 한칸씩 이동 - # 안정성 1 감소
        # 만약 그 사람이 n보다 크면 아웃
    if len(personList) > 0:
        personLen = len(personList)
        for eachIdx in range(personLen-1, -1, -1):

            curPersonPose = personList[eachIdx]

            if curPersonPose == n-1: # n-1에 있으면
                personList.pop()
                continue

            if eachIdx == len(personList)-1: # n-1에 있을 수는 없음. 근데 마지막 사람이면 안정성만 확인
                if(stableList[curPersonPose+1] > 0):
                    personList[eachIdx] += 1
                    stableList[personList[eachIdx]] -= 1
                continue

            # 마지막이 아닌 사람들
            nextPersonPose = personList[eachIdx+1]
            if(nextPersonPose-curPersonPose > 1 and stableList[curPersonPose+1] > 0):
                personList[eachIdx] += 1
                stableList[personList[eachIdx]] -= 1

    print('aft people step')
    print('st', stableList)
    print('pe', personList)
    print()

    # 0번 무빙워크 칸에 사람 없고 안정성 0 아니면 personList에 사람 추가 (0 추가)
    # 안정성 1 감소
    if len(personList) > 0:
        if personList[0] > 0 and stableList[0] > 0:
            personList.insert(0, 0)
            stableList[0] -= 1
    else:
        if stableList[0] > 0:
            personList.insert(0, 0)
            stableList[0] -= 1

    print('aft new person')
    print('st', stableList)
    print('pe', personList)
    print()

    # 안정성이 0인 칸이 k개 이상이라면 과정을 종료
    cnt = 0
    for i in stableList:
        if(i == 0):
            cnt += 1

        if(cnt == k):
            keepGo = False
            break

print(turn)
```
