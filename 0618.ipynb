```python
# 풍선 터뜨리기 - 프로그래머스
# 못풀고 다른 사람 풀이 참고함

def solution(a):

    # 처음에 낸 아이디어
    # -92 -68 -70 => 안됨
    # -92 -68 -33 => 됨
    # 각 숫자를 기준으로 왼쪽 min 오른쪽, min에 자기보다 작은게 최대 하나만 있어야 됨
    ## 이것도 괜찮은 아이디어인데 절대 안되네... 주인공 바뀔 때마다 오른쪽에서 min 구하는게 시간이 오래걸림

    # 다른 사람 풀이
    # X, min(a) 가 있을 때 0 번째에서 X 자리까지 원소들 중 X가 가장 작은지 확인
    # min, X 가 있을 때 X 자리 ~ 끝 중 X가 제일 작은지 확인


    minNum = min(a)
    minIdx = a.index(minNum)
    temNum = 1000000000
    cnt = 0
    for idx in range(0,minIdx):
        if(temNum > a[idx]):
            temNum = a[idx]
            cnt += 1

    temNum = 1000000000
    for idx in range(len(a)-1, minIdx, -1):
        if(temNum > a[idx]):
            temNum = a[idx]
            cnt += 1

    return cnt+1 # minNum도 추가
```
