```python
# 나무 섭지 (softeer)
# https://softeer.ai/practice/7726#pop_user







## 남우보다 아무 유령이나 먼저 exit 가면 되는거임 (중간에 만나고 뭐고 다 필요없음)
## 출입구에서 가장 가까운 유령과 출입구 사이의 거리 and 현재 남우와 출입구 사이의 거리 비교하면 됨

n, m = list(map(int, input().split()))

# mat에는 벽, 빈공간만
# 남우 visited mat
# 남우 pose
# 출구 pose
# 유령 각각 pose

mat = [['.' for c in range(m)] for r in range(n)]
namewooVisited = [[False for c in range(m)] for r in range(n)]
namwooPose = []
exitPose = []

ghostMat = [[0 for c in range(m)] for r in range(n)]
# ghostPoseList = []

ghostExist = False
for r in range(n):
    tem = list(input())
    for c in range(m):
        if(tem[c] == 'N'):
            namwooPose = [r,c]
            tem[c] = '.'
        elif(tem[c] == 'D'):
            exitPose = [r,c]
            tem[c] = '.'
        elif(tem[c] == 'G'):
            # ghostPoseList.append([r,c])
            ghostExist = True
            ghostMat[r][c] = 1
            tem[c] = '.'

    mat[r] = tem[:]

# for r in range(n):
#     print(mat[r])




dys = [0,-1,0,1]
dxs = [1,0,-1,0]

def in_range(y,x):
    return 0<=y and y<n and 0<=x and x<m

# exit pose에서 가장 가까운 유령 pose 구하기
def findGhost():

    queue = [exitPose[:]]
    visited = [[False for j in range(m)] for i in range(n)]
    visited[exitPose[0]][exitPose[1]] = True

    candList = []

    while(queue):
        cy, cx = queue.pop(0)

        for dy, dx in zip(dys, dxs):
            ny, nx = cy+dy, cx + dx
            if(not in_range(ny, nx) or visited[ny][nx] == True):
                continue

            visited[ny][nx] = True
            queue.append([ny,nx])

            if(ghostMat[ny][nx] == 1):
                candList.append([ny,nx])

        if(candList):
            break


    # 남우한테서 가장 가까운 애만 남기기
    candList.sort(key=lambda x : (abs(namwooPose[0] - x[0]) + abs(namwooPose[1] - x[1])))
    return candList[0]


def namwooSearch(source, target):

    queue = []
    # visited = [[namewooVisited[i][j] for j in range(m)] for i in range(n)]
    visited = [[False for j in range(m)] for i in range(n)]

    curMat = [[-1 for j in range(m)] for i in range(n)]

    cy, cx = source
    queue.append(source[:])
    visited[cy][cx] = True
    curMat[cy][cx] = 0

    while(queue):

        cy, cx = queue.pop(0)

        for dy, dx in zip(dys, dxs):
            ny, nx = cy+dy, cx + dx

            if(not in_range(ny, nx) or visited[ny][nx] == True or mat[ny][nx] == '#'):
                continue

            queue.append([ny,nx])
            visited[ny][nx] = True
            curMat[ny][nx] = curMat[cy][cx] + 1

            if([ny,nx] == target):
                return curMat[ny][nx]

    return 0


nearGhost = []
if ghostExist == True:
    nearGhost = findGhost()[:]

namwooCnt = namwooSearch(namwooPose, exitPose)

# print(nearGhost)
# print(namwooCnt)

if(namwooCnt == 0):
    print('No')
elif namwooCnt > 0 and ghostExist:
    cmp1 = namwooCnt
    cmp2 = abs(nearGhost[0]-exitPose[0]) + abs(nearGhost[1]-exitPose[1])
    if(cmp1 < cmp2): # 남우가 더 가까우면
        print('Yes')
    else:
        print('No')
else:
    print('Yes')
```
