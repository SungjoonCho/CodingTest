```python
# https://www.codetree.ai/training-field/frequent-problems/problems/cube-rounding-again/description?page=2&pageSize=20
# 정육면체 한번 더 굴리기 - 코드트리



n, m = list(map(int, input().split()))
mat = [[0 for i in range(n)] for j in range(n)]

for i in range(n):
    mat[i] = list(map(int, input().split()))

# 주사위 표현
# 위, 정면, 오른쪽만 표현
curCube = [1,2,3]


# 주사위 이동 방향
# 우 하 좌 상
curDir = 0
dys = [0,1,0,-1]
dxs = [1,0,-1,0]

def in_range(y,x):
    return 0<=y and y<n and 0<=x and x<n

def updateCube():
    global curCube

    # 앞뒤좌우로 굴렸을 때 케이스 나눠서 구현
    # 주사위 위, 정면, 오른쪽 숫자 업데이트
    temCube = curCube[:]
    if(curDir == 0): # 우
        curCube[0] = 7 - temCube[2]
        curCube[2] = temCube[0]

    elif (curDir == 1):  # 하
        curCube[1] = temCube[0]
        curCube[0] = 7 - temCube[1]

    elif (curDir == 2):  # 좌
        curCube[0] = temCube[2]
        curCube[2] = 7 - temCube[0]

    elif (curDir == 3):  # 상
        curCube[0] = temCube[1]
        curCube[1] = 7 - temCube[0]

def calScore():

    curVal = mat[cy][cx]
    # print(cy, cx, curVal)

    queue = [[cy,cx]]
    visited = [[False for i in range(n)] for j in range(n)]
    visited[cy][cx] = True
    cnt = 0

    while(queue):
        curY, curX = queue.pop(0)
        cnt += 1
        # print(curY, curX)

        for dy, dx in zip(dys, dxs):
            ny = curY + dy
            nx = curX + dx

            if(not in_range(ny,nx) or mat[ny][nx] != curVal or visited[ny][nx] == True):
                continue

            queue.append([ny,nx])
            visited[ny][nx] = True

    # print(curVal, cnt, '=====')
    return curVal * cnt




totalScore = 0
cy, cx = 0, 0 #  현재 위치
for round in range(m):

    # 현재 방향으로 굴릴 수 있는지 확인
        # in range 벗어나면 이동 방향 반대로 업데이트
    if(not in_range(cy+dys[curDir], cx+dxs[curDir])):
        curDir = (curDir+2)%4

    # 현재 방향으로 한번 굴리기
        # 현재 주사위 위치 업데이트
        # 현재 주사위의 모습 업데이트
    cy, cx = cy+dys[curDir], cx+dxs[curDir]
    updateCube()
    # print(curDir, curCube)

    # bfs해서 점수 계산
    score = calScore()
    totalScore += score

    # print(score)

    # 현재 칸이랑 숫자 비교해서 방향 업데이트
    curVal = mat[cy][cx]
    curBottom = 7 - curCube[0]
    if(curBottom > curVal):
        curDir = (curDir+1)%4
    elif(curBottom < curVal):
        curDir = (curDir - 1) % 4



print(totalScore)

```
