```python
# 징검다리 (소프티어)
# https://jie0025.tistory.com/441 참고

n = int(input())
stoneList = list(map(int, input().split()))
dpList = [1] * n

for i in range(1, n):
    find = 0
    for j in range(i):
        if(stoneList[j] < stoneList[i]):
            find = max(find, dpList[j])

    dpList[i] = find+1
# print(dpList)
print(max(dpList))
```
