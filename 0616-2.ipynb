```python

# 벽 부수고 이동하기 4 - 백준
# https://www.acmicpc.net/problem/16946


# sameBlockMat에 같은 그룹 숫자 표시
# recordMat에 각 그룹의 최대값 표시

# virusMat 1인 칸에 대해서 네 방향 sameBlockMat 값 확인 and 아직 그룹 숫자가 없으면 bfs 진행, recordMat 값 합산


n, m = list(map(int,input().split()))
virusMat = [[0 for i in range(m)] for j in range(n)]

for r in range(n):
    virusMat[r] = list(map(int, list(input())))

sameBlockMat = [[-1 for i in range(m)] for j in range(n)]
recordMat = [[0 for i in range(m)] for j in range(n)]

# for r in range(n):
#     print(virusMat[r])

dys = [0,-1,0,1]
dxs = [1,0,-1,0]

curBlockNum = 0

def in_range(y,x):
    return 0<=y and y<n and 0<=x and x<m

def bfs(arg_y, arg_x):
    global recordMat, sameBlockMat, curBlockNum

    queue = []
    queue.append([arg_y, arg_x])
    visited = [[arg_y, arg_x]]
    sameBlockMat[arg_y][arg_x] = curBlockNum

    # sameBlockMat에 같은 블록 표시
    while(queue):
        cy, cx = queue.pop()
        for dy, dx in zip(dys, dxs):
            ny, nx = cy + dy, cx + dx

            # 원래 [ny,nx] in visited 였는데 sameBlockMat[ny][nx] >= 0 으로 고치니까 시간초과 해결됨
            # in을 매번 하는게 상당히 시간 많이 잡아먹나봄
            if(not in_range(ny,nx) or sameBlockMat[ny][nx] >= 0 or virusMat[ny][nx] == 1):
                continue

            queue.append([ny,nx])
            visited.append([ny,nx])
            sameBlockMat[ny][nx] = curBlockNum

    # recordMat에 최대값 표시
    maxNum = len(visited)
    for cy, cx in visited:
        recordMat[cy][cx] = maxNum

    # 업데이트
    curBlockNum += 1


totalMat = [[0 for i in range(m)] for j in range(n)]

# 같은 블럭인지 확인 후 숫자 기록
for r in range(n):
    for c in range(m):

        # virusMat[r][c] 1 이면
            # totalMap 기본값 = 1
            # 상하좌우 for문
                # sameBlockMat[r][c]이 -1인 경우
                    # 맵핑 후 값 리턴 후 합산
                # 아니면
                    # record 값 읽어와서 totalMap에 합산
        # 0이면 맵핑만
            # 상하좌우 for문
                # sameBlockMat[r][c]이 -1인 경우 맵핑

        if(virusMat[r][c] == 0):
            continue

        curCnt = 1
        sameBlockList = []
        for dy, dx in zip(dys, dxs):
            ny, nx = r + dy, c + dx
            if (not in_range(ny, nx) or virusMat[ny][nx] == 1 or sameBlockMat[ny][nx] in sameBlockList):
                continue

            if(sameBlockMat[ny][nx] == -1):
                bfs(ny,nx)

            curCnt += recordMat[ny][nx]

            sameBlockList.append(sameBlockMat[ny][nx])

        totalMat[r][c] = (curCnt%10)


for r in range(n):
    print("".join(map(str,totalMat[r])))
```
