```python
# 프로그래머스(등굣길)


def solution(m, n, puddles):
	answer = 0

	mat = [[0 for i in range(n+1)] for j in range(m+1)]
	pudMat = [[False for i in range(n+1)] for j in range(m+1)]

	mat[1][1] = 1
	for r,c in puddles:
	    pudMat[r][c] = True

	for r in range(m+1):
	    for c in range(n+1):
	        if(pudMat[r-1][c] == False):
	            mat[r][c] += mat[r-1][c]
	        if(pudMat[r][c-1] == False):
	            mat[r][c] += mat[r][c-1]


	return mat[-1][-1] % 1000000007
```
