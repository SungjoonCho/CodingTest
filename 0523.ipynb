```python
# 꼬리잡기놀이 - 코드트리


## 순서
# 한칸 이동 - 어디 방향으로 이동할지
# 공 던지고
# 점수 획득
# 방향 바꾸기


n, m, k = list(map(int, input().split()))
mat = [[0 for i in range(n)] for j in range(n)]
teamIdx_mat = [[0 for i in range(n)] for j in range(n)]
teamInfo = []

for r in range(n):
    mat[r] = list(map(int, input().split()))

    # 1 위치 저장
    for c in range(n):
        if (mat[r][c] == 1):
            head = [r, c]
            totalBody = [head]
            pathList = []
            temScore = 0
            totalScore = 0

            teamInfo.append([totalBody, pathList, temScore, totalScore])
            teamIdx_mat[r][c] = len(teamInfo)


## 정보 관리 - [[head, body, tail], [경로 리스트 (현재 가야하는 위치에서 제일 우선인 것부터)], 공 맞았으면 임시 점수 0 , 총합 점수 0]
# 1의 위치 저장, 1부터 4를 먹어가면서 경로 저장 (4인 곳만 저장), body랑 tail 저장

def in_range(y, x):
    return 0 <= y and y < n and 0 <= x and x < n


dys = [0, -1, 0, 1]
dxs = [1, 0, -1, 0]

# 하우상좌
dys = [1, 0, -1, 0]
dxs = [0, 1, 0, -1]

for teamIdx in range(len(teamInfo)):
    eachTeam = teamInfo[teamIdx]
    head = eachTeam[0][0]
    # cy, cx = head
    visited = [head]
    queue = [head]

    # body랑 tail, 경로 추가
    tail = []
    while queue:
        cy, cx = queue.pop(0)

        for dy, dx in zip(dys, dxs):
            ny, nx = cy + dy, cx + dx
            if (not in_range(ny, nx) or mat[ny][nx] == 0 or [ny, nx] in visited):
                continue

            teamIdx_mat[ny][nx] = teamIdx + 1

            if (mat[ny][nx] == 2):
                teamInfo[teamIdx][0].append([ny, nx])
                visited.append([ny, nx])
            elif (mat[ny][nx] == 3):
                tail = [ny, nx]
                visited.append([ny, nx])
                continue
            elif (mat[ny][nx] == 4):
                teamInfo[teamIdx][1].append([ny, nx])
                visited.append([ny, nx])

            queue.append([ny, nx])

    teamInfo[teamIdx][0].append(tail)

# for i in teamInfo:
#     print(i)
# print()
# print(teamInfo)

## 공 던지기 리스트 만들기 (시작지점, dy, dx)
ballList = []
for dy, dx in zip(dys, dxs):
    start = []
    if dy == 1 and dx == 0:  # 하
        start = [0, 0]
        go_y, go_x = 0, 1
    elif dy == 0 and dx == 1:  # 우
        start = [n - 1, 0]
        go_y, go_x = -1, 0
    elif dy == -1 and dx == 0:  # 상
        start = [n - 1, n - 1]
        go_y, go_x = 0, -1
    elif dy == 0 and dx == -1:  # 좌
        start = [0, n - 1]
        go_y, go_x = 1, 0

    for i in range(n):
        ballList.append([[start[0] + dy * i, start[1] + dx * i], go_y, go_x])

# print(ballList)


ball_idx = 0

for round in range(k):

    if(round > 0):
        ## 방향 바꾸기
        # 임시 점수 0보다 크면
        # [head, body, tail]을 역순으로 다시 저장, 경로 리스트 역순으로 저장
        for teamIdx in range(len(teamInfo)):
            temScore = teamInfo[teamIdx][2]
            if (temScore > 0):
                teamInfo[teamIdx][0] = teamInfo[teamIdx][0][::-1]
                teamInfo[teamIdx][1] = teamInfo[teamIdx][1][::-1]

                teamInfo[teamIdx][2] = 0


    # print(teamInfo[2][1])

    ## 경로 리스트에서 제일 앞에 있는걸로 한 칸 이동 - head body tail 업데이트, 경로 리스트 맨 뒤에 tail 있던 곳 추가
    # mat 업데이트 (head body tail에 해당하는 숫자로 업데이트)
    for teamIdx in range(len(teamInfo)):


        # 경로 전체가 몸인 경우
        if (len(teamInfo[teamIdx][1]) == 0):
            # print(teamInfo[teamIdx][0])


            old_head = teamInfo[teamIdx][0][0]
            old_tail = teamInfo[teamIdx][0][-1]

            mat[old_head[0]][old_head[1]] = 2
            mat[old_tail[0]][old_tail[1]] = 2


            new_head = teamInfo[teamIdx][0][-1]
            new_tail = teamInfo[teamIdx][0][-2]

            mat[new_head[0]][new_head[1]] = 1
            mat[new_tail[0]][new_tail[1]] = 3



            new_tail = teamInfo[teamIdx][0].pop()
            teamInfo[teamIdx][0].insert(0, new_tail)

            # print(teamInfo[teamIdx][0])

        else:

            # head
            # 기존 head 처리
            head = teamInfo[teamIdx][0][0]
            mat[head[0]][head[1]] = 2

            # 새 head 처리
            goto = teamInfo[teamIdx][1].pop(0)

            teamInfo[teamIdx][0].insert(0, goto)  # 새 head 지정
            mat[goto[0]][goto[1]] = 1

            # tail
            tail = teamInfo[teamIdx][0].pop()
            mat[tail[0]][tail[1]] = 4
            teamInfo[teamIdx][1].append(tail)

            tail = teamInfo[teamIdx][0][-1]
            mat[tail[0]][tail[1]] = 3

    # print(teamInfo)

    # for i in range(n):
    #     print(mat[i])
    # print()

    ## 공 던지기 - idx 1씩 증가시키면서 던질 r이나 c 지정
    # for문으로 mat 확인
    # 임시 점수 0 이면 거기에 값 넣어주기
    # [head, body, tail]에서 몇번째 값인지 확인해서 값 구하고 넣어주기

    # 되면 - 둘다 True이면 break

    start, go_y, go_x = ballList[ball_idx]
    # print(start, go_y, go_x)

    for i in range(n):
        cy, cx = start[0] + go_y * i, start[1] + go_x * i

        # if (not in_range(cy, cx)):
        #     continue

        if (mat[cy][cx] > 0 and mat[cy][cx] < 4):
            teamidx = teamIdx_mat[cy][cx] - 1

            if (teamInfo[teamidx][2] == 0):
                targetIdx = teamInfo[teamidx][0].index([cy, cx]) + 1
                teamInfo[teamidx][2] += (targetIdx ** 2)
                break ##### 한 명 있으면 바로 끝

    # print(teamInfo)

    # 임시 점수를 총합에 합산
    for teamIdx in range(len(teamInfo)):
        teamInfo[teamIdx][-1] += teamInfo[teamIdx][2]
        # print(teamInfo[teamIdx][2])

    ball_idx += 1
    if (ball_idx == len(ballList)):
        ball_idx = 0



    # break

totalSum = 0
for each in teamInfo:
    # print(each[-1])
    totalSum += each[-1]

print(totalSum)
```
