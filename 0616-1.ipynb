```python

# 바이러스 실험 - 코드트리
# https://www.codetree.ai/training-field/frequent-problems/problems/virus-experiment/description?page=3&pageSize=20

n, m, k = list(map(int, input().split()))

nu_add_mat = [[0 for i in range(n)] for j in range(n)]
nu_db_mat = [[5 for i in range(n)] for j in range(n)]
virus_pose_mat = [[[] for i in range(n)] for j in range(n)]

for i in range(n):
    nu_add_mat[i] = list(map(int, input().split()))

for i in range(m):
    v_r,v_c,age = list(map(int, input().split()))
    virus_pose_mat[v_r-1][v_c-1].append(age)

# print(virus_pose_mat)

dys = [0,-1,-1,-1,0,1,1,1]
dxs = [1,1,0,-1,-1,-1,0,1]

def in_range(y,x):
    return 0<=y and y<n and 0<=x and x<n

for each_round in range(k):
    # 양분 섭취
        # sort
        # nu_db_mat - 나이 , 나이 +1
        # 양분 부족하면 pop 후 따로 save (나이 나누기 2 - 소수점 버리기)

    # for r in range(n):
    #     print(virus_pose_mat[r])
    #
    # print()
    #
    # for r in range(n):
    #     print(nu_db_mat[r])
    #
    # print()

    for r in range(n):
        for c in range(n):
            if(len(virus_pose_mat[r][c]) == 0):
                continue
            virus_pose_mat[r][c].sort()

            save = 0
            popCnt = 0
            saveList = []
            for each_idx in range(len(virus_pose_mat[r][c])):
                each_age = virus_pose_mat[r][c][each_idx]
                if(each_age <= nu_db_mat[r][c]):
                    nu_db_mat[r][c] -= each_age
                    virus_pose_mat[r][c][each_idx] += 1
                    saveList.append(virus_pose_mat[r][c][each_idx])
                else:
                    for rem in virus_pose_mat[r][c][each_idx:]:
                        save += int(rem/2)
                    break

            virus_pose_mat[r][c] = saveList[:]
            # 양분 + save 총합
            if(save > 0):
                nu_db_mat[r][c] += save

    # 번식 진행 - 8칸에 나이 +1 바이러스 추가 (in_range)
    for r in range(n):
        for c in range(n):
            if(len(virus_pose_mat[r][c]) == 0):
                continue

            for each_idx in range(len(virus_pose_mat[r][c])):
                age = virus_pose_mat[r][c][each_idx]
                if(age % 5 != 0):
                    continue

                for dy, dx in zip(dys,dxs):
                    ny = r + dy
                    nx = c + dx
                    if(not in_range(ny, nx)):
                        continue

                    virus_pose_mat[ny][nx].append(1)


    # 양분 전체 추가
    for r in range(n):
        for c in range(n):
            nu_db_mat[r][c] += nu_add_mat[r][c]

total = 0
for r in range(n):
    for c in range(n):
        total += len(virus_pose_mat[r][c])

print(total)
```
